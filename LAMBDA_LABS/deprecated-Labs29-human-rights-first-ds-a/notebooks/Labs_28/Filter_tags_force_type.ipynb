```python
import pandas as pd
import spacy

df = pd.read_csv('all_sources_geoed.csv',usecols=[1,2,3,4,5,7,8,11,12])
npl = spacy.load('en_core_web_sm')

df.head(15)
```

```python
tags = []
for x in df['tags']:
    y = x.strip('['']')
    tags.append(y)
u_tags = set()
for i in tags:
    j = i.strip('"')
    k = j.split(",")
    for x in k:
        u_tags.add(x)
        
new = []
for x in u_tags:
    x = x.strip()
    x = x.strip('\'')
    new.append(x)

u_new = set(new)
again = []
for i in u_new:
    again.append(i)

drop = ['','hide-badge', 'non-protest', 'horse', 'bike','person-with-disability', 'journalist','child','medic', 'elderly',
              'legal-observer','arrest','body-cam','abuse-of-power', 'pregnant','property-destruction','protester', 
              'politician','bystander','racial-profiling','lgbtq+', 'drive','homeless','conceal', 'sexual-assault',
       'inhumane-treatment']
for x in drop:
    
     again.remove(x)
print(again)
```

```python
def fixer(q):
    temp = []
    reject = []

    j = q.strip('['']')
    x = j.split(",")
    for g in x:
        z = g.strip()
        z = z.strip('\'')  
        if z in again:
            temp += [z]
        
    return temp
            
df['force_type'] = df['tags'].apply(fixer)
df = df.drop('tags',axis = 1)
df.head()
```

```python
df.to_csv('all_sources_geoed_forcetyped.csv')
```
