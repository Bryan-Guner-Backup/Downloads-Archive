```python
import pandas as pd
df = pd.read_csv("all_sources_geoed.csv")
df. head()
```

```python
import ast
tag_words = set()
for i in df['tags']:
    x = ast.literal_eval(i)
    for q in x:
        tag_words.add(q)
tag_list = []        
for i in tag_words:
    if i != '':
        tag_list.append(i)
print(tag_list)
```

```python
drop_list = ['bystander', 'person-with-disability', 'pregnant','body-cam', 'drive', 'elderly','non-protest','homeless','bike',
             'celebrity', 'horse','protester', 'medic', 'legal-observer', 'dog', 'journalist', 'vehicle', 'racial-profiling',
             'lgbtq+', 'property-destruction', 'child', 'politician', 'conceal', 'hide-badge']

for x in drop_list:
    if x in tag_list:
        tag_list.remove(x)
print(tag_list)
```
